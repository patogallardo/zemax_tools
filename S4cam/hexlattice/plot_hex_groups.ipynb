```python
from matplotlib.patches import RegularPolygon
from matplotlib.collections import PatchCollection
```

```python
fig, ax = plt.subplots(figsize=[8, 8])
patches = []
p = RegularPolygon([0, 0], 6, 0.8, 0, color="C1")
patches.append(p)
p = RegularPolygon([3,3], 6, 0.8, 0, color="C1")
patches.append(p)
collection = PatchCollection(patches, fc=["C1", "C2"])
ax.add_collection(collection)
plt.xlim([-5,5])
plt.ylim([-5,5])
```
